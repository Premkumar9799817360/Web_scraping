```python


import requests
from bs4 import BeautifulSoup as bs
# url = "https://www.bookswagon.com/promo-best-seller/now-trending/E5B93FF87A87"
url="https://www.bookswagon.com/promo-best-seller/now-trending/E5B93FF87A87?sid=419753"
response = requests.get(url)
soup = bs(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
books = soup.find_all(class_="booktitle font-weight-bold")
print(books)
```

```python
books_data = []
books = soup.find_all(class_="booktitle font-weight-bold")
for book in books:
    name = book.get_text(strip=True)
    print(book.get_text())
    books_data.append(name)
print(books_data)
```
