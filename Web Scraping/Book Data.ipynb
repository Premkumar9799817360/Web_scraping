```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.bookswagon.com/"
response = requests.get(url)
```

```python
soup = bs(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
print(soup.find_all(class_="booktitle font-weight-bold"))
```

```python
books_data = []
books = soup.find_all(class_="booktitle font-weight-bold")
for book in books:
    name = book.get_text(strip=True)
#     print(book.get_text())
    books_data.append(name)
print(books_data)
```

```python
for i in books_data:
    print(i)
```

```python
print(soup.select("author authortextcolor"))
```

```python
print(soup.find_all(class_="author authortextcolor"))
```

```python
author_name = []
author = soup.find_all(class_="author authortextcolor")
for name in author:
    i = name.get_text(strip=True)
    author_name.append(i)
print(author_name)
```

```python
for author in author_name:
    print(author)
```

```python
print(soup.find_all(class_="actualprice themecolor  font-weight-bold"))
```

```python
prices = soup.find_all("span", class_="actualprice")
```

```python
print(prices)
```

```python
price_data = []
price= soup.select("span.actualprice")
# print(price_elements)
for i in price:
    name = i.get_text()
    price_data.append(name)
#     print(name)
print(price_data)
```

```python
price_data.remove('\n\n')
```
