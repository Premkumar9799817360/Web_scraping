```python
import requests
from bs4 import BeautifulSoup 
```

```python
url = input("Enter an url for web scraping: ")
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
tag = soup.h4
print(tag)
```

```python
heading = soup.find_all('h4')
for head in heading:
    print(head.get_text())
```
