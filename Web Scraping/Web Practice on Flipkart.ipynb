```python
import requests
from bs4 import BeautifulSoup

# Replace this URL with the URL of the webpage you want to scrape
url = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=8138f03e-cbdd-4384-8613-f6aacb54754a"

# Send a GET request to the URL
response = requests.get(url,headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all elements with class containing "title" or "heading"
    title_and_heading_elements = soup.find_all(class_=lambda value: value and ("title" in value or "heading" in value))
    
    # Extract and print the text content of each element
    for element in title_and_heading_elements:
        print(element.get_text())
else:
    print("Failed to retrieve the webpage.")
```
